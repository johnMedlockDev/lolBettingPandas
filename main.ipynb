```python
# read in new df

df = pd.read_csv("./csvs/cleaned/summary.csv", delimiter = ',')
```

```python
# rename key column

df.rename(columns={df.columns[0]: "key"}, inplace=True)
```

```python
# replace NAs in player columns

df.update(df[["dragons","barons",'towers',"total cs","firstblood"]].fillna(0))
df = df.fillna('Team')
```

```python
#fix columns and row formatting

df.columns =map(str.capitalize,df.columns)
df['Position'] = df['Position'].str.capitalize()
df['Player'] = df['Player'].str.capitalize()
df['Gamelength'] = df['Gamelength'] /60
df['Firstblood'] = df["Firstblood"].astype(float)
df.loc[df['Gamelength'] < 30 , 'Gamelength'] = 2
df.loc[df['Gamelength'] > 30 , 'Gamelength'] = 0
```

```python
# Score Calculation
df["Score"] = (df["Kills"] * 3) + (df["Assists"] * 2) + (df["Total cs"] * 0.02) + (df["Towers"] *1) +(df["Gamelength"])+ (df["Firstblood"]*2)+ (df["Dragons"] * 2)+ (df["Barons"]*3)- (df["Deaths"] * 1)
```

```python
df1 = df[['Player', "Score","Position"]]
for team in teams:
    print("====================")
    print(team)
    print(df1.where(df["Team"] ==team).groupby("Player").mean().round(2).sort_values(by=['Score'], ascending=False))
    for position in positions:
        print()
        print(position)
        print( df1.where(df["Team"] ==team).where(df["Position"] == position).groupby("Player").mean().round(2).sort_values(by=['Score'], ascending=False))
```
